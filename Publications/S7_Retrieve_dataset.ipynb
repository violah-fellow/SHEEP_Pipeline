```python
import duckdb
import pandas as pd
from datetime import datetime

# today
today = datetime.today().strftime("%y%m%d")
```

```python
# Connect to database
db = duckdb.connect('publications.db')
```

```python
# Show tables in database
db.sql("SHOW TABLES")
```

```python
# get labelled data
df = db.sql("SELECT * FROM publications_labelled").df()
```

```python
df
```

```python
# save data as csv
df.to_csv(f"data/Publications_data_labelled_{today}.csv")
```

```python
db.close()
```
